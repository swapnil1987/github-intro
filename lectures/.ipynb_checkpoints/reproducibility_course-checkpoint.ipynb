# What is Markdown?

- A **lightweight markup language** created in 2004
- Plain text that converts to HTML (and PDF, Word, etc.)
- Human-readable even without rendering
- Used by: GitHub, Jupyter, R Markdown, Obsidian, and more

## Why Markdown for Research?

âœ… Plain text = Future-proof  
âœ… Version control friendly  
âœ… Focus on content, not formatting  
âœ… Universal support

# Markdown: Headers

## Syntax:

```markdown
# Heading 1
## Heading 2
### Heading 3
#### Heading 4
```

## Result:

# Heading 1
## Heading 2
### Heading 3
#### Heading 4

# Markdown: Text Formatting

## Syntax:

```markdown
*italic* or _italic_
**bold** or __bold__
***bold and italic***
~~strikethrough~~
`inline code`
```

## Result:

*italic* or _italic_  
**bold** or __bold__  
***bold and italic***  
~~strikethrough~~  
`inline code`

# Markdown: Lists

## Unordered:
```markdown
- Item 1
- Item 2
  - Nested item
```

## Ordered:
```markdown
1. First step
2. Second step
3. Third step
```

## Task Lists:
```markdown
- [x] Completed task
- [ ] Incomplete task
```

# Markdown: Links & Images

## Links:
```markdown
[Link text](https://example.com)
[Link with title](https://example.com "Hover text")
```

## Images:
```markdown
![Alt text](path/to/image.png)
![Logo](https://example.com/logo.png "Optional title")
```

## Reference Style:
```markdown
[my paper][1]

[1]: https://doi.org/10.1234/example
```

# What is Git?

## A Distributed Version Control System

- Created by Linus Torvalds in 2005
- Tracks changes to files over time
- Enables collaboration without conflicts
- Works offline; syncs when ready

# Essential Git Commands

## Setup & Init:
```bash
git config --global user.name "Your Name"
git config --global user.email "you@example.com"
git init                    # Start a new repo
git clone <url>             # Copy an existing repo
```

## Daily Workflow:
```bash
git status                  # What's changed?
git add <file>              # Stage changes
git add .                   # Stage all changes
git commit -m "message"     # Save to history
```

# Essential Git Commands (cont.)

## Viewing History:
```bash
git log                     # See commit history
git log --oneline           # Compact view
git diff                    # See unstaged changes
git diff --staged           # See staged changes
```

## Syncing with Remote:
```bash
git remote add origin <url> # Connect to GitHub
git push                    # Upload commits
git pull                    # Download changes
```

# Branching & Merging

## Create and switch branches:
```bash
git branch experiment       # Create branch
git checkout experiment     # Switch to it
# OR combined:
git checkout -b experiment  # Create and switch
```

## Merge back:
```bash
git checkout main           # Go back to main
git merge experiment        # Merge experiment into main
```

**Best Practice:** One branch per feature/experiment

# Writing Good Commit Messages

## âŒ Bad:
```
git commit -m "fixed stuff"
git commit -m "updates"
git commit -m "asdfasdf"
```

## âœ… Good:
```
git commit -m "Add normalization to preprocessing pipeline"
git commit -m "Fix off-by-one error in simulation loop"
git commit -m "Update README with installation instructions"
```

**Format:** Imperative mood, ~50 characters, explain *why*

# Choosing a License

## Common Open Source Licenses:

| License | Permissions | Conditions |
|---------|-------------|------------|

# Common Git Problems & Solutions

## "I committed to the wrong branch!"
```bash
git checkout correct-branch
git cherry-pick <commit-hash>
```

## "I need to undo my last commit!"
```bash
git reset --soft HEAD~1    # Keep changes staged
git reset --hard HEAD~1    # Discard changes âš ï¸
```
